```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
import os

def plot_power_spectrum(audio_path, label):
    # Load the audio file
    y, sr = librosa.load(audio_path)

    # Compute the Short-Time Fourier Transform (STFT)
    D = np.abs(librosa.stft(y))**2

    # Compute the power spectrum
    power_spectrum = librosa.amplitude_to_db(D, ref=np.max)

    # Plot the power spectrum
    plt.figure(figsize=(10, 6))
    librosa.display.specshow(power_spectrum, sr=sr, x_axis='time', y_axis='log', cmap='magma')
    plt.colorbar(format='%+2.0f dB')
    plt.title(label)
    plt.xlabel('Time (s)')
    plt.ylabel('Frequency (Hz)')
    plt.show()
```

```python
audio_dir = 'music'


audio_files = [f for f in os.listdir(audio_dir) if f.endswith('.wav')]

for audio_file in audio_files:
    audio_path = os.path.join(audio_dir, audio_file)
    plot_power_spectrum(audio_path, audio_file)
```

```python
import matplotlib.pyplot as plt

# Data
models = ['KMeans', 'KNN', 'RandomForest', 'SVC', 'NN']
train_accuracy = [70, 93.5, 100, 100, 100]  # Training set accuracy
test_accuracy = [52, 68, 48, 62, 90.6]       # Testing set accuracy


# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.35
index = range(len(models))

bars1 = ax.bar(index, train_accuracy, bar_width, label='Précision de l\'entraînement', color='cornflowerblue')
bars2 = ax.bar([i + bar_width for i in index], test_accuracy, bar_width, label='Précision des tests', color='lightcoral')

ax.axhline(y=90, color='blue', linestyle='--', label='Benchmark (90%)')
ax.axhline(y=70, color='red', linestyle='--', label='Benchmark (70%)')

ax.set_xlabel('Modèles')
ax.set_ylabel('Précision (%)')
ax.set_title('Précision de l’entraînement et des tests de différents modèles')
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(models)
ax.legend()

# Adding the actual numbers on top of the bars
def add_labels(bars):
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval + 1, round(yval, 1), ha='center', va='bottom')

add_labels(bars1)
add_labels(bars2)

plt.tight_layout()
plt.show()
```
